```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#line chart
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 3]
plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```

```python
## 2. Bar Chart
latihan bar chart dengan membandingkan dua dataset
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]
values2 = [30.0, 35.0, 25.0, 25.0]

x = np.arange(len(categories)) #posisi x untuk tiap kategori
width = 0.35 # lebar bar

plt.bar(x + width/2, values2, width, label='set 2')
plt.bar(x - width/2, values, width, label='set 1')

plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
## 3. Pie Chart
Latihan pie chart proporsi kategori dengan data persentase
```

```python
label = ['Kategori a', 'Kategori b', 'Kategori c', 'Kategori d']
ukuran = [25, 25, 30, 20]

plt.pie(ukuran, labels=label, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
## 4. Histogram
latihan membuat histogram dari satu data random (Histogram ke1)
latihan membuat histogram dari dua data random untuk melihat perbandingan penyebaran frequensinya (Histogram ke2)
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=35, edgecolor='black')
plt.title('Histogram')
plt.xlabel('values')
plt.ylabel('frequensi')
plt.show()

```

```python
#buat dua dataset random

data_a = np.random.randn(1000)
data_b = np.random.randn(1000) + 2

plt.hist(data_a, bins=35, alpha=0.6, color="blue", edgecolor="black", label="Dataset A")
plt.hist(data_b, bins=35, alpha=0.6, color="red", edgecolor="black", label="Dataset A")

plt.title("Perbandingan Dua Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
## 5. Scatter Plot
latihan scatter plot 2 variabel data random (Scatter ke1)
Latihan scatter plot 3 variabel dengan warna (Scatter ke2)
```

```python
x = np.random.rand(50)
y = 2*x+1+0.1+np.random.rand(50)

plt.scatter (x,y)
plt.title('Scatter')
plt.xlabel('Values')
plt.ylabel('Frequensi')
plt.show()
```

```python
np.random.seed(42)
nama_siswa = [f"siswa{i}" for i in range(1,11)]

nilai_Matematika = np.random.randint(60, 100, 10)
nilai_IPA = np.random.randint(60, 100, 10)
nilai_Bahasa = np.random.randint(60, 100, 10)

plt.scatter(nilai_Matematika, nilai_IPA, c=nilai_Bahasa, cmap="viridis",
            s=nilai_Bahasa*5, alpha=0.8, edgecolor="Black")

for i, nama in enumerate(nama_siswa):
  plt.text(nilai_Matematika[i]+0.5, nilai_IPA[i]+0.5, nama)

plt.title("Scatter plot 3 variabel")
plt.xlabel("Nilai Matematika (X)")
plt.ylabel("Nilai IPA (Y)")
plt.colorbar(label="Nilai Bahasa")
plt.show()
```

```python
## 6. BoxPlot
Latihan membuat boxplot dengan tiga dataset random

## Analisis
kemungkinan outlier paling banyak ada di dataset B
```

```python
np.random.seed(42)
data_A = np.random.normal(50, 10, 100)
data_B = np.random.normal(60, 15, 100)
data_C = np.random.normal(55, 5, 100)

plt.boxplot([data_A, data_B, data_C], labels=["A", "B", "C"])
plt.title("Box Plot 3 Dataset Random")
plt.ylabel("Nilai")
plt.show()
```

```python
## 7. Heatmap
latihan membuat heatmap dataset random dengan visualisasi dengan 3 variasi yaitu dasar, anotasi, dan colormap
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
data = {
    "A" : np.random.randn(50),
    "B" : np.random.randn(50),
    "C" : np.random.randn(50),
    "D" : np.random.randn(50),
}
df = pd.DataFrame(data)
corr = df.corr()

sns.heatmap(corr)
plt.title("Heatmap Korelasi (Dasar)")
plt.show()

sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Heatmap Korelasi (Dengan Anotasi)")
plt.show()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi (coolwarm)")
plt.show()

```

```python
## 8. Area Chat
-Latihan membuat Area Chart dengan data random penjualan produk
-Area chart ini menunjukkan perbandingan dua data penjualan produk setiap bulan
-Produk B menunjukkan range area penjualan lebih tinggi dari produk A
```

```python
x = np.arange(1, 13)
Sales_A = np.random.randint(20, 50, size=12)
Sales_B = np.random.randint(30, 60, size=12)

df = pd.DataFrame({
    "Bulan": x,
    "Produk_A": Sales_A,
    "Produk_B": Sales_B
})

df.set_index("Bulan").plot(kind="area", alpha=0.6)

plt.title("Area Chart Penjualan Produk A & B Per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.show()
print(df)
```
